````python
df['Count'] = 1
df['City Title'] = df['City'].str.title()  # 'clifton park', 'new york', 'rochester' not capitalized
df.head()
````

````python
station_count = df[['City Title', 'Count']]
top_10=station_count.groupby('City Title', as_index=False) \
                    .sum() \
                    .sort_values(['Count'], ascending=False) \
                    .head(10)
top_10
````

````python
ny_stations = df[df['City Title'] == 'New York']
ny_stations.head()
````

````python
df[df['City Title'] == 'Queens']
````

````python
df[df['City Title'] == 'Astoria']
````

````python
nyc_stations = df[                         # You can also use .isin(['New York', ...]) alternatively
    (df['City Title'] == 'New York') |     # Why must use Bitwise Operator instead of Logical Operator?
    (df['City Title'] == 'Bronx') | 
    (df['City Title'] == 'Queens') |       # Queens is weird with its City names (has 'Astoria', 'Flushing', etc.)
    (df['City Title'] == 'Brooklyn') |
    (df['City Title'] == 'Staten Island')
]
                
nyc_stations.head()
````

````python
nyc_station_count = nyc_stations[['City Title', 'Count']]
nyc=nyc_station_count.groupby('City Title', as_index=False) \
                     .sum() \
                     .sort_values(['Count'], ascending=False) \

nyc
````

````python
### your code here

fig, ax = plt.subplots(figsize=(12, 12))

ny_counties.plot(
    ax=ax,
    facecolor='lightgrey',
)

df.plot(
    ax=ax,
    title = 'Electric Vehicle Charging Stations in New York',
    x = 'Longitude',
    y = 'Latitude',
    kind = 'scatter',
)

plt.show()
````

#### (Optional) Do any additional or other Maps in cell(s) below;

````python
### your optional code here
import folium
from folium.plugins import MarkerCluster

m = folium.Map(
    location=[
        df['Latitude'].mean(),
        df['Longitude'].mean(),
    ],
    zoom_start = 6,
)

mc = MarkerCluster()

for i, row in df.iterrows():  #for row in df.itertuples():
    mc.add_child(
        folium.Marker(
            location=[row.Latitude, row.Longitude],
            popup=row['Station Name']  # popup=row.ZIP
        )
    )
    
    m.add_child(mc)

m
````

````python
m.save('index.html')
````
